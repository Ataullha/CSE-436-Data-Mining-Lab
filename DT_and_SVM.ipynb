```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
X = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=81)
```

```python
# from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
# model = SVC(kernel='rbf',C=1,gamma='scale',random_state=81)
model = DecisionTreeClassifier(criterion='entropy', max_depth=2, random_state=81)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
plot_tree(model)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, classification_report
```

```python
# actual_value, predicted_value
```

```python
cm = confusion_matrix(y_test, model.predict(X_test), labels=model.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
disp.plot()
```
