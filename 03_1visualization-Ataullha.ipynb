```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Ataullha/CSE-436-Data-Mining-Lab/refs/heads/main/titanic_train.csv")
```

```python
df.head(2)
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df['age'].isnull().sum()
```

```python
data = df[['age']]
```

```python
sns.catplot(data=data,kind='box')
```

```python
data.describe()
```

```python
sns.catplot(data=data,kind='violin')
```

```python
sns.catplot(x='sex',y='survived',kind='bar',data=df)
```

```python
sns.catplot(x='sex',y='survived',hue='pclass',kind='bar',data=df)
```

```python
sns.pairplot(df)
```

```python
sns.jointplot(df[['age','fare']])
```
